```python
df = pd.read_csv('drug_repur/entities.tsv',sep='\t',header=None)  
all_item = set()
for index, row in df.iterrows():
  all_item.add(row[1])
```

```python
food = pd.read_csv('drug_repur/food.csv',sep='|') 
food_subset = []
for index, row in food.iterrows():
  if row[1] in all_item:
    #print(row[1])
    food_subset.append(row[1])
```

```python
drug = pd.read_csv('drug_repur/drug.csv') 
drug_subset = []
for index, row in drug.iterrows():
  if row[2] in all_item:
    #print(row[1])
    drug_subset.append(row[2])
```
